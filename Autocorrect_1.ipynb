```python
import re
import string
from collections import Counter
import numpy as np
import pandas as pd

```

```python
def read_corpus(filename):
  with open(filename, "r") as file:
    lines = file.readlines()
    words = []
    for line in lines:
      words += re.findall(r'\w+', line.lower())

  return words
```

```python
words = read_corpus("./vocab2.txt")
print(f"There are {len(words)} total words in the corpus")
```

```python
vocabs = set(words)
print(f"There are {len(vocabs)} unique words in the vocabulary")
```

```python
word_counts = Counter(words)
print(word_counts["the"])
```

```python
total_word_count = float(sum(word_counts.values()))
word_probas = {word: word_counts[word] / total_word_count for word in word_counts.keys()}
```

```python
def split(word):
  return [(word[:i], word[i:]) for i in range(len(word) + 1)]
```

```python
print(split("passed"))
```

```python
def delete(word):
  return [l + r[1:] for l,r in split(word) if r]
```

```python
print(delete("passed"))
```

```python
def swap(word):
  return [l + r[1] + r[0] + r[2:] for l, r in split(word) if len(r)>1]
```

```python
print(swap("passed"))
```

```python
string.ascii_lowercase
```

```python
def replace(word):
  letters = string.ascii_lowercase
  return [l + c + r[1:] for l, r in split(word) if r for c in letters]
```

```python
print(replace("passed"))
```

```python
def insert(word):
  letters = string.ascii_lowercase
  return [l + c + r for l, r in split(word) for c in letters]
```

```python
print(insert("passed"))
```

```python
def edit1(word):
  return set(delete(word) + swap(word) + replace(word) + insert(word))
```

```python
print(edit1("passed"))
```

```python
def edit2(word):
  return set(e2 for e1 in edit1(word) for e2 in edit1(e1))
```

```python
print(edit2("passed"))
```

```python
def correct_spelling(word, vocabulary, word_probabilities):
  if word in vocabulary:
    print(f"{word} is already correctly spelt")
    return

  suggestions = edit1(word) or edit2(word) or [word]
  best_guesses = [w for w in suggestions if w in vocabulary]
  return [(w, word_probabilities[w]) for w in best_guesses]
```

```python
word = "smle"
corrections = correct_spelling(word, vocabs, word_probas)

if corrections:
  print(corrections)
  probs = np.array([c[1] for c in corrections])
  best_ix = np.argmax(probs)
  correct = corrections[best_ix][0]
  print(f"{correct} is suggested for {word}")
```

```python
# Load the dataset
# Replace 'your_dataset.csv' with the actual path to your dataset
data = pd.read_csv('misspelled.csv')

# Assume the dataset has columns named 'misspelled' and 'correct'
test_words = data['input'].tolist()
ground_truth = data['label'].tolist()

```

```python
# Define the model class to use the dataset
class DatasetAutocorrectModel:
    def __init__(self, data):
        # Create a dictionary for fast lookup of corrections
        self.correction_dict = dict(zip(data['input'], data['label']))

    def predict(self, word):
        # Return the corrected word if found, otherwise return the original word
        return self.correction_dict.get(word, word)
```

```python
# Initialize the model with the dataset
model = DatasetAutocorrectModel(data)
```

```python
# Function to generate predictions using the autocorrect model
def generate_predictions(model, test_words):
    predictions = []
    for word in test_words:
        corrected_word = model.predict(word)  # Use the model to correct the word
        predictions.append(corrected_word)
    return predictions

# Generate predictions using your model
predictions = generate_predictions(model, test_words)
```

```python
# Function to calculate accuracy
def calculate_accuracy(predictions, targets):
    correct_predictions = sum([1 for pred, target in zip(predictions, targets) if pred == target])
    accuracy = correct_predictions / len(targets)
    return accuracy
```

```python
# Calculate and print the accuracy
accuracy = calculate_accuracy(predictions, ground_truth)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
class SpellChecker(object):

  def __init__(self, corpus_file_path):
    with open(corpus_file_path, "r") as file:
      lines = file.readlines()
      words = []
      for line in lines:
        words += re.findall(r'\w+', line.lower())

    self.vocabs = set(words)
    self.word_counts = Counter(words)
    total_words = float(sum(self.word_counts.values()))
    self.word_probas = {word: self.word_counts[word] / total_words for word in self.vocabs}

  def _level_one_edits(self, word):
    letters = string.ascii_lowercase
    splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deletes = [l + r[1:] for l,r in splits if r]
    swaps = [l + r[1] + r[0] + r[2:] for l, r in splits if len(r)>1]
    replaces = [l + c + r[1:] for l, r in splits if r for c in letters]
    inserts = [l + c + r for l, r in splits for c in letters]

    return set(deletes + swaps + replaces + inserts)

  def _level_two_edits(self, word):
    return set(e2 for e1 in self._level_one_edits(word) for e2 in self._level_one_edits(e1))

  def check(self, word):
    candidates = self._level_one_edits(word) or self._level_two_edits(word) or [word]
    valid_candidates = [w for w in candidates if w in self.vocabs]
    return sorted([(c, self.word_probas[c]) for c in valid_candidates], key=lambda tup: tup[1], reverse=True)
```

```python
checker = SpellChecker("./vocab2.txt")
```
